```python
import pandas as pd

# Load the data from the new CSV file
data = pd.read_pickle("../../../data/data.pkl")

# Display the first few rows of the data
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the missing values in each column
missing_values = data.isnull()

# Plot a heatmap of the missing values
plt.figure(figsize=(10, 6))
sns.heatmap(~missing_values, cbar=False, yticklabels=False)
plt.title("Data Completeness Heatmap (Black: Data, White: Missing)", fontsize=13)
plt.show()
```
